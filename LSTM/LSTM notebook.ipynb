```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
# tokenizer.word_counts
tokenizer.word_index
len(tokenizer.word_index)
# tokenizer.count
```

```python
for sentence in data.split('\n'):
  print(sentence)
```

```python
list=[]
for sentence in data.split('\n'):
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    list.append(tokenized_sentence[:i+1])
```

```python
list
```

```python
# need to add zero padding
```

```python
length=0
for i in list:
  if len(i)>length:
    length=len(i)
length
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
df=pad_sequences(list,length,padding='pre')
```

```python
df
```

```python
x=df[:,:-1]
x.shape
```

```python
y=df[:,-1]
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=632)
```

```python
y
```

```python
y.shape
```

```python
# now lstm model

# 1)embedding layer
# 2)lstm layer
# 3)dense layer
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(632,100,input_length=29))
model.add(LSTM(150))
model.add(Dense(632,activation='softmax'))
# same no of dense neuron as no.of words,because every word will give a probability due to softmax function
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42
                                                 )
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```
